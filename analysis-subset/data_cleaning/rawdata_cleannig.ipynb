```python
import pandas as pd
import json
from IPython.display import display
import os
```

```python
folder_path = 'full_data'
files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]
dataframes = []

for file in files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    if 'messages' in df.columns:
        messages_df = df[df['messages'].notna() & (df['messages'].str.strip() != '')]
        if 'data' in messages_df.columns:
            messages_df = messages_df.drop(columns=['data'])

        dataframes.append(messages_df)

        if 'count' in messages_df.columns:
            count_values = messages_df['count'].value_counts()

            count_values_df = pd.DataFrame({
                'count_value': count_values.index,
                'occurrences': count_values.values
            })

            print (f'{file}Total submission: {count_values_df["occurrences"].sum()}')

        else:
            print("The 'count' column was not found in the messages DataFrame.")

raw_combined = pd.concat(dataframes, ignore_index=True)
display(raw_combined)
```

```python
# find missing conversations, ensure all conversations are runned
folder_path = 'full_data'
files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

all_counts = set()

for file in files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)

    if 'messages' in df.columns:
        messages_df = df[df['messages'].notna() & (df['messages'].str.strip() != '')]

        if 'data' in messages_df.columns:
            messages_df = messages_df.drop(columns=['data'])

        if 'count' in messages_df.columns:
            all_counts.update(messages_df['count'].dropna().unique())
        else:
            print(f"The 'count' column was not found in {file}.")
        

all_counts = set(map(int, all_counts))

expected_counts = set(range(0,1456))

missing_counts = expected_counts - all_counts

print(f"Missing count values: {sorted(missing_counts)}")
print(len(missing_counts))
```

```python
# num of rows processed
final_df = pd.DataFrame()
count = 0

for file in files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    for idx, msg in df['messages'].items():
        if pd.isna(msg):
            continue

        try:
            data = json.loads(msg)
        except json.JSONDecodeError:
            print(f"Skipping row {idx} due to JSON decoding error.")
            continue
        data_df = pd.DataFrame(data)

        data_df['transcript_id'] = df.loc[idx, 'conversation_id']

        data_df['count'] = df.loc[idx, 'count']

        final_df = pd.concat([final_df, data_df], ignore_index=True)

        count += 1

    # Save the cleaned DataFrame to a new CSV file
final_df.to_csv(f'./all_cleaned.csv', index=False)
print(f"Total valid rows processed: {count}")
```

```python
# import pandas as pd

file_path = "all_cleaned.csv"
df = pd.read_csv(file_path)
df.rename(columns={'participent_id': 'PID'}, inplace=True)

# Number of unique transcripts
num_transcripts = df['transcript_id'].nunique()
print(f"Number of annotated transcripts: {num_transcripts}")

# Annotations per participant
annotations_per_participant = df.groupby('PID')['new_topic'].nunique().reset_index()
annotations_per_participant.rename(columns={'new_topic': 'annotation_count'}, inplace=True)

# Average annotations
print(annotations_per_participant['annotation_count'].sum() / 1529)

# Participants with fewer than 6 annotations
few_annotations = annotations_per_participant[annotations_per_participant['annotation_count'] < 6]
few_pids = few_annotations['PID'].tolist()
print(f"\nNumber of participants with fewer than 6 annotations: {len(few_annotations)}")
display(few_annotations)

# Filter df to only those participants
few_df = df[df['PID'].isin(few_pids)]

# All other participants (6 or more annotations)
other_pids = annotations_per_participant[annotations_per_participant['annotation_count'] >= 6]['PID'].tolist()
other_df = df[df['PID'].isin(other_pids)]

few_counts = set(few_df['count'].unique())
other_counts = set(other_df['count'].unique())

# Remove any counts that appear in both groups
unique_few_counts = few_counts - other_counts
unique_few_counts = sorted(int(x) for x in unique_few_counts)

print("\nUnique count values for participants with fewer than 6 annotations (excluding shared ones):")
print(sorted(unique_few_counts))
print(f"Number of unique count values: {len(unique_few_counts)}")
```
